```python
#Ejercicio 1

#a. Define los métodos:
# 1. dormir, que emitirá el mensaje ‘ZZZZzzzzzzzzz’
# 2. hablar, que emitirá el mensaje ‘bla bla blaaaaa’
# 3. contar, que emitirá el mensaje ‘1,2,3,4,…’ (números enteros positivos, mayores a 0, hasta el tope que le
# indiques)
# 4. emitir, que emitirá el nombre dado al Robot y la edad en función de la fecha de puesta en marcha.

#b. El programa deberá pedir al usuario:
# 1. si desea crear un Robot, en el caso de que su respuesta sea afirmativa, instanciamos el objeto.
# 2. que le dé un nombre, id, fecha de puesta en marcha.
# 3. que le indique si quiere que emita su nombre.
# 4. que le indique si quiere que hable.
# 5. que le indique si quiere que vaya a dormir.
# 6. que le indique si quiere escucharlo contar pero debe decirle hasta qué número

import datetime
from datetime import date

class Robot:
    nombre = ""
    id = ""
    fecha = date.today() # fecha de puesta en marcha

    def __init__(self, nombre, id, fecha):
        self.nombre = nombre
        self.id = id
        self.fecha = fecha
    
    def __str__(self):
        cadena = 'Nombre: ' + self.nombre + ', id: ' + str(self.id) + ', fecha: ' + str(self.fecha)
        return cadena

    def dormir(self):
        print("ZZZZzzzzzzzzz")

    def hablar(self):
        print("bla bla blaaaaa")

    def contar(self):
        n = int(input("Ingrese un numero y se mostrara los enteros positivos desde el 1 hasta ese numero: "))
        for i in range(n):
            print(i+1)

    def emitir(self):
        fecha_comienzo = self.fecha
        today = date.today()
        
        print('Nombre: ' + self.nombre + ', edad: ' + str(today.year - fecha_comienzo.year - ((today.month, today.day) < (fecha_comienzo.month, fecha_comienzo.day))))


#Main:

respuesta = input("Desea crear un robot? S / N")

if(respuesta == "S"):
    nombre = input("Ingrese nombre del robot")
    id = int(input("Ingrese el id del robot"))

    anio = int(input("Ingrese el año de puesta en marcha del robot"))
    mes = int(input("Ingrese el mes de puesta en marcha del robot"))
    dia = int(input("Ingrese el dia de puesta en marcha del robot"))
    fecha = datetime.date(anio, mes, dia)

    robot1 = Robot(nombre, id, fecha)
    print(robot1)
    robot1.hablar()
    robot1.contar()
    robot1.emitir()
```

```python
#Ejercicio 2

#a. Crea la clase Actividad. Esta clase hereda de Robot.
#b. Define los atributos: nombre y edad.
#c. Redefine el método emitir para que emita todos los datos.
#d. Puedes definir una o más actividades o trabajos para el robot?. Si es así define al menos uno

class Actividad(Robot):
    edad = 0

    def __init__(self, nombre, edad):
        self.nombre = nombre
        self.edad = edad

    def emitir(self):
        print('Nombre: ' + nombre + ' Edad: ' + str(edad))
    
    def jugar(self):
        print("wiiii")

nombre = input("Ingrese nombre del robot")
edad = int(input("edad"))
robot2 = Actividad(nombre, edad)

robot2.hablar()
robot2.emitir()
robot2.jugar()
```

```python
#Ejercicio 3.
# a. Crea la clase HijoDeRobot. Esta clase hereda de Robot y Actividad
# b. Define los atributos: nombre y edad, según la fecha de puesta en marcha
# c. Redefine el método emitir para que emita los datos del hijo
# d. Si la edad es superior a 3 años puede tener una actividad y emitirla, en caso contrario debe emitir una
# excepción: ‘No tiene actividad, todavía está aprendiendo!’ (ver pdf errores, excepciones…..)

class HijoDeRobot(Actividad):

    def __init__(self, nombre, edad):
        self.nombre = nombre
        today = date.today()
        self.edad = int(today.year - edad.year - ((today.month, today.day) < (edad.month, edad.day)))

    def emitir(self):
        print('Nombre: ' + nombre + ' Edad: ' + str(self.edad))
    
    def realizar_actividad(self):
        if(self.edad > 3):
            self.jugar()
        else:
            raise Exception("No tiene actividad, todavía está aprendiendo!")

nombre = input("Ingrese nombre del robot")
anio = int(input("Ingrese el año de puesta en marcha del robot"))
mes = int(input("Ingrese el mes de puesta en marcha del robot"))
dia = int(input("Ingrese el dia de puesta en marcha del robot"))
fecha = datetime.date(anio, mes, dia)

robot3 = HijoDeRobot(nombre, fecha)

robot3.hablar()
robot3.emitir()
robot3.realizar_actividad()
```
