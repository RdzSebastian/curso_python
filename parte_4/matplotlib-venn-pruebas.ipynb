```python
import matplotlib.pyplot as plt
from matplotlib_venn import venn2

#uso una variable para poder hacer mas acciones sobre el diagrama y quede mas legible
diagram = venn2((1,1,1), set_colors=("#FF5744", "#68FF33"), alpha=1)
#el id 11 es la interseccion, el 01 es el primer valor y el 10 es el segundo 
diagram.get_label_by_id("11").set_text("Interseccion")
plt.show
```
