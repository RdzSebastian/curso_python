```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# Ejercicio 1
L1 = ['a','b','c','d','e','f','g','z'] 
L2 = [('a','b', 4), ('a','d', 5), ('a','g', 2), ('z','e', 3), ('z','f', 2), ('z','c', 1), ('e','d', 2), ('c','b', 5), ('c','f', 1), ('e','f', 1), ('d','g', 1), ('g','b', 1)]

G = nx.Graph()
G.add_nodes_from(L1)
G.add_weighted_edges_from(L2)  

pos_nodes['a'] = [0.500, 0.200]
pos_nodes['b'] = [0.400, 0.100]
pos_nodes['c'] = [0.200, 0.100]
pos_nodes['d'] = [0.400, 0.300]
pos_nodes['e'] = [0.200, 0.300]
pos_nodes['f'] = [0.200, 0.200]
pos_nodes['g'] = [0.400, 0.200]
pos_nodes['z'] = [0.100, 0.200]

nx.draw_networkx_nodes(G, pos_nodes, nodelist=list(G.nodes), node_color="pink", node_size=1000)
pos.update(pos_nodes)

nx.draw_networkx_labels(G, pos, font_size=10, font_family='sans-serif')
nx.draw_networkx_edges(G, pos, edge_color='lightblue', width=4)

labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)
plt.axis('off')
plt.show()

print("Ruta mas corta usando el algoritmo de Dijkstra entre z y a: ", nx.algorithms.dijkstra_path(G, 'z', 'a'))
```

```python
# Ejercicio 2
L1 = ['a','b','c','d','e','f','g','h','i','j','k','l','m','n'] 
L2 = [
    ('a','b', 6), 
    ('a','c', 9), 
    ('a','d', 3), 
    ('b','e', 12), 
    ('b','c', 10), 
    ('c','e', 3), 
    ('c','f', 2), 
    ('c','d', 4),
    ('d','f', 15), 
    ('d','g', 10), 
    ('e','h', 4),
    ('e','i', 11),
    ('e','f', 8),
    ('f','i', 10),
    ('f','g', 9),
    ('g','j', 13),
    ('h','k', 20),
    ('h','l', 11),
    ('h','i', 7),
    ('i','l', 11),
    ('i','j', 15),
    ('j','l', 11),
    ('j','m', 9),
    ('k','n', 6),
    ('k','l', 13),
    ('l','n', 12),
    ('l','m', 5),
    ('m','n', 5)]

G = nx.Graph()
G.add_nodes_from(L1)
G.add_weighted_edges_from(L2)  

pos_nodes['a'] = [0.100, 0.200]
pos_nodes['b'] = [0.150, 0.300]
pos_nodes['c'] = [0.150, 0.200]
pos_nodes['d'] = [0.150, 0.100]
pos_nodes['e'] = [0.200, 0.300]
pos_nodes['f'] = [0.200, 0.200]
pos_nodes['g'] = [0.200, 0.100]
pos_nodes['h'] = [0.250, 0.300]
pos_nodes['i'] = [0.250, 0.200]
pos_nodes['j'] = [0.250, 0.100]
pos_nodes['k'] = [0.300, 0.300]
pos_nodes['l'] = [0.300, 0.200]
pos_nodes['m'] = [0.300, 0.100]
pos_nodes['n'] = [0.350, 0.200]

nx.draw_networkx_nodes(G, pos_nodes, nodelist=list(G.nodes), node_color="black", node_size=1000)
pos.update(pos_nodes)

nx.draw_networkx_labels(G, pos, font_size=10, font_family='sans-serif', font_color = 'blue')
nx.draw_networkx_edges(G, pos, edge_color='blue', width=4)

labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels, font_color = 'red')
plt.axis('off')
plt.show()

print("Ruta mas corta usando el algoritmo 'camino mas corto' entre a y n: ", nx.algorithms.shortest_path(G, 'a', 'n'))

print("Radio: " + str(nx.radius(G, e=None, usebounds=False)))
print("Diametro: " + str(nx.diameter(G, e=None, usebounds=False)))
print("Excentricidad: " + str(nx.eccentricity(G)))
print("Centro: " + str(nx.center(G, e=None, usebounds=False)))
print("Perisferia: " + str(nx.periphery(G, e=None, usebounds=False)))
print("Densidad: " + str(nx.density(G)))
```
